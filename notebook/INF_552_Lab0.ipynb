#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)

````python

# Create an empty dictionary to store the data
data_dict = {}

# Iterate over the rows of the DataFrame
for index, row in df.iterrows(): #The iterrows() function in Python is a method used with a Pandas DataFrame to iterate over the rows. It generates an iterator that yields each row of the DataFrame as a pair consisting of: the index of the row and a series representing the row's data
    # Iterate over the columns of the current row
    for column, value in row.items():
        # If the column header is not yet in the dictionary, create a new list
        if column not in data_dict:
            data_dict[column] = []
        # Append the value to the corresponding list
        data_dict[column].append(value)

# Print the dictionary (optional)
print(data_dict)
````

Read the documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html#pandas.DataFrame and create a dataframe using pd.DataFrame from the dictionary created in the iterrows() exercise. Change the header to "a", "b", "c", ... at creation time.

````python
pd.DataFrame(data_dict) #recreate the dataframe from the dictionary we created above using the "DataFrame" function
````
